```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cfb22.csv')
```

```python
df.head()
```

```python
df['Percent Defense Plays'] = df['Def Plays'] / (df['Off Plays'] + df['Def Plays'])
```

```python
percdef = df.groupby('Percent Defense Plays')
percdef['Win-Loss'].max()
```

```python
df.head()
```

```python
df[['Win', 'Loss']] = df['Win-Loss'].str.split('-', expand=True)
print(df)
```

```python
df.head()
```

```python
df["Win"] = df["Win"].apply(int)
```

```python
df['Win Percent'] = df['Win'] / df['Games']
```

```python
print(df['Win Percent'])
```

```python
df["Win Percent"] = df["Win Percent"].apply(float)
df["Percent Defense Plays"] = df["Percent Defense Plays"].apply(float)
```

```python
df.plot.scatter(x='Percent Defense Plays', y='Win Percent', alpha=0.6)
plt.xlabel('Percent Defense Plays')
plt.ylabel('Win Percent')
```

```python
df.plot.scatter(x='Off Yards/Play', y='Win Percent', alpha=0.6)
plt.xlabel('Offensive Yards per Play')
plt.ylabel('Win Percent')
```
